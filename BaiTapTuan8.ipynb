```python
# W7A18
def dividenconquer(a):
    n = len(a)
    if n==1:
        even = []
        odd = []
        if a[0]%2==0:
            even.append(a[0])
        else:
            odd.append(a[0])
        return (even, odd)
    pair1 = dividenconquer(a[:n//2])
    pair2 = dividenconquer(a[n//2:])
    return conquer(pair1, pair2)
    
def conquer(pair1, pair2):
    even = []
    i = j = 0
    while i<len(pair1[0]) and j<len(pair2[0]):
        if pair1[0][i]<pair2[0][j]:
            even.append(pair1[0][i])
            i += 1
        else:
            even.append(pair2[0][j])
            j += 1
    while i<len(pair1[0]):
        even.append(pair1[0][i])
        i += 1
    while j<len(pair2[0]):
        even.append(pair2[0][j])
        j += 1
    odd = []
    i = j = 0
    while i<len(pair1[1]) and j<len(pair2[1]):
        if pair1[1][i]>pair2[1][j]:
            odd.append(pair1[1][i])
            i += 1
        else:
            odd.append(pair2[1][j])
            j += 1
    while i<len(pair1[1]):
        odd.append(pair1[1][i])
        i += 1
    while j<len(pair2[1]):
        odd.append(pair2[1][j])
        j += 1
    return even, odd
a = list(map(int, input().split()))
print(*dividenconquer(a)[0], *dividenconquer(a)[1])
```

```python
# W7A19
def check(a, b, increasing = True):
    if increasing:
        return a<b
    else:
        return a>b
a = list(map(int, input().split()))
b = list(map(int, input().split()))
i = 0
j = 0
res = []
increasing = a[0]<a[1]
while i<len(a) and j<len(b):
    if check(a[i], b[j], increasing):
        res.append(a[i])
        i += 1
    else:
        res.append(b[j])
        j += 1
while i<len(a):
    res.append(a[i])
    i += 1
while j<len(b):
    res.append(b[j])
    j += 1
print(*res)
```

```python
# W7A27
def check_brackets(x, y):
    if x=="(" and y==")":
        return True
    elif x=="[" and y=="]":
        return True
    elif x=="{" and y=="}":
        return True
    else:
        return False
s = input()
bracket_dct = {"<": ">", "(": ")", "[": "]", "{": "}"}
stack = []
flag = True
for b in s:
    if len(stack)==0:
        if b in bracket_dct.values():
            flag = False
            break
        else:
            stack.append(b)
    else:
        if b in bracket_dct.values():
            if bracket_dct[stack[-1]] == b:
                stack.pop()
            else:
                flag = False
                break
        else:
            stack.append(b)
flag = flag and len(stack)==0
if flag:
    print("YES")
else:
    print("NO")
```

```python
# W7A30
def lis(a, end, memo=[]):
    if memo == []:
        memo = [[]]*len(a)
    if memo[end] != []:
        return memo[end].copy()
    elif end==0:
        memo[0] = [a[0]]
        return [a[0]]
    elif end < len(a)-1:
        longest = [a[end]]
        for i in range(end):
            current = liq(a, i, memo)
            if a[end] > current[-1]:
                current += [a[end]]
                if len(current) > len(longest):
                    longest = current.copy() 
        memo[end] = longest.copy()
        return longest
    else:
        longest = []
        for i in range(end):
            current = liq(a, i, memo)
            if a[end] > current[-1]:
                current += [a[end]]
            if len(current) > len(longest):
                longest = current.copy() 
        return longest

a = list(map(int, input().split()))
print(*lis(a, len(a)-1))
```

```python
# W7A31
def dynamic_min_id(a, start):
    res = start
    for i in range(start+1,len(a)):
        if a[i] < a[res]:
            res = i
            candidate = a[i]
        elif a[i] == a[res] and a[res] == 0:
            res = i
    return res
n = input()
digits = list(map(int, list(n)))
dct = {}
duplicate = False
swapped = False
skip = False
for i in range(len(digits)):
    dct[digits[i]] = dct.get(digits[i],0)+1
    if dct[digits[i]]>1:
        duplicate = True
    if not skip:
        min_id = dynamic_min_id(digits, i)
    if digits[i] == digits[min_id]:
        if i > min_id:
            skip = True
        else:
            skip = False
    else:
        digits[i], digits[min_id] = digits[min_id], digits[i]
        swapped = True
        break
if not swapped and not duplicate:
        digits[-1], digits[-2] = digits[-2], digits[-1]
print(*digits, sep="")
```
