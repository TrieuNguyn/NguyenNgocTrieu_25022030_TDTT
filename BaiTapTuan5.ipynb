```python
# W5A1
def max_of_two(a, b):
    if a >= b:
        return a
    else:
        return b

def main():
    a, b = map(int, input().split())
    print(max_of_two(a, b))

main()
```

```python
# W5A2
def swap(a, b):
    return (b, a)

def main():
    a, b = map(int, input().split())
    print(swap(a, b))

main()
```

```python
# W5A3
def isprime(n):
    if n < 2:
        return False
    elif n == 2:
        return True
    else:
        i = 2
        while i*i<=n:
            if n%i==0:
                return False
            i += 1
        return True
def main():
    n = int(input())
    print(isprime(n))

main()
```

```python
# W5A4
from math import sqrt

def isperfect(n):
    s = 1
    i = 2
    while i*i <= n:
        if n%i==0:
            s += i
        if n/i > i:
            s += n/i
        i += 1
    return n==s

def main():
    n = int(input())
    print(isperfect(n))

main()
```

```python
# W5A5
def search(lst, k):
    for i in range(len(lst)):
        if lst[i] == k:
            return i
    return -1

def main():
    lst = list(map(int, input().split()))
    k = int(input())
    print(search(lst, k))

main()
```

```python
# W5A6
def factorial(n):
    assert n>=0
    if n <= 1:
        return 1
    else:
        return n*factorial(n-1)

def main():
    n = int(input())
    print(factorial(n))

main()
```

```python
# W5A7
def calculator(exp):
    if exp[2] == "+":
        return round(float(exp[0])+float(exp[1]), 2)
    elif exp[2] == "-":
        return round(float(exp[0])-float(exp[1]), 2)
    elif exp[2] == "*":
        return round(float(exp[0])*float(exp[1]), 2)
    elif exp[2] == "/" and exp[1] != 0:
        return round(float(exp[0])/float(exp[1]), 2)
        
def main():
    exp = input().split()
    print(calculator(exp))

main()
```

```python
# W5A8
def hamming_distance(a, b):
    count = 0
    while True:
        if a%2 != b%2:
            count += 1
        a = a//2
        b = b//2
        if a==0 and b==0:
            break  
    return count
        
def main():
    a, b = map(int, input().split())
    print(hamming_distance(a ,b))

main()
```

```python
# W5A9
def sum_digits(n):
    s = 0
    while True:
        s += n%10
        n = n//10
        if n==0:
            break
    return s

def main():
    n = int(input())
    print(sum_digits(n))

main()
```

```python
# W5A9
from string import ascii_lowercase 
def isisomorphic(a, b):
    if len(a) != len(b): # eliminate the case in which the lengths of the strings are unmatched
        return False

    a, b = a.lower(), b.lower() # for consistency

    def idx(c): # convert the ASCII code of a character to its index in ascii_lowercase for convenience
        return ord(c)-ord('a')

    # let the mapping be from string a to string b
    A = list(ascii_lowercase) # A, now being the alphabet array, would be modified to reflect the mapping: a[i] -> A[idx(a[i])] = b[i]
    flag_a = [False]*len(A) # flag if character ascii_lowercase[i] is mapped from
    flag_b = [False]*len(A) # flag if character ascii_lowercase[i] is mapped to
    for i in range(len(a)):
        if not flag_a[idx(a[i])] and not flag_b[idx(b[i])]: # if character a[i] and b[i] are both unmapped
                A[idx(a[i])] = b[i] # mapping
                flag_a[idx(a[i])] = True # reflagging as they are mapped to each other
                flag_b[idx(b[i])] = True
        elif flag_a[idx(a[i])] and flag_b[idx(b[i])]: # if character a[i] and b[i] are both mapped 
            if A[idx(a[i])] != b[i]: # check if a[i]->b[i] is consistent with the mapping in process
                return False
        else: # case in which one is mapped while the other is not
            return False
    return True    
def main():
    a, b = input().split()
    print(isisomorphic(a, b))

main()
```

```python
# W5A9 v2
def isisomorphic(a, b):
    if len(a) != len(b):
        return False
    a, b = a.lower(), b.lower() 
    dcta = {}
    dctb = {}
    for i in range(len(a)):
        if dcta.get(a[i],"") == "":
            dcta[a[i]] = b[i]
        elif dcta[a[i]] != b[i]:
            return False
        if dctb.get(b[i],"") == "":
            dctb[b[i]] = a[i]
        elif dctb[b[i]] != a[i]:
            return False
    return True
    
def main():
    a, b = input().split()
    print(isisomorphic(a, b))

main()
```
