```python
def merge_sort(a):
    if len(a) == 1:
        return a
    mid = len(a)//2
    l = merge_sort(a[:mid])
    r = merge_sort(a[mid:])
    return merge(l, r)

def merge(l, r):
    i = 0
    j = 0
    res = []
    while True:
        if l[i] < r[j]:
            res.append(l[i])
            i += 1
        else:
            res.append(r[j])
            j += 1
        if i==len(l) or j==len(r):
            break
    res.extend(l[i:])
    res.extend(r[j:])
    return res

a = list(map(int, input().split()))
print(merge_sort(a))
```
