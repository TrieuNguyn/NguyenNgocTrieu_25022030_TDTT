```python
# W5A9

def isisomorphic(a, b):
    if len(a) != len(b):
        return False
    a, b = a.lower(), b.lower() 
    dcta = {}
    dctb = {}
    for i in range(len(a)):
        if dcta.get(a[i],"") == "":
            dcta[a[i]] = b[i]
        elif dcta[a[i]] != b[i]:
            return False
        if dctb.get(b[i],"") == "":
            dctb[b[i]] = a[i]
        elif dctb[b[i]] != a[i]:
            return False
    return True
    
def main():
    a, b = input().split()
    print(isisomorphic(a, b))

main()
```
