```python
# W9A2
def merge(sorted_a, row, i):
    res = sorted_a[:i+1]
    p = i+1
    q = 0
    while p<len(sorted_a) and q<len(row):
        if sorted_a[p]<row[q]:
            res.append(sorted_a[p])
            p += 1
        else: 
            res.append(row[q])
            q += 1
    res.extend(sorted_a[p:])
    res.extend(row[q:])
    return res
n, m = map(int, input().split())
a = []
for _ in range(n):
    a.append(list(map(int, input().split())))
k = int(input())
a_flattened = []
for i in range(n):
    a_flattened = merge(a_flattened, a[i], i)
print(a_flattened[k-1])
```

```python
# W9A3
def dynamic_min_id(a, start):
    small = a[start]
    idx = start
    for i in range(start+1, len(a)):
        if small > a[i]:
            idx = i
            small = a[i]
        elif small == a[i] and small == 0:
            idx = i
    return idx

n = input()
digits = list(n)
digits = list(map(int, digits))
cnt = {}
duplicate = False
swapped = False
for i in range(len(digits)-1):
    cnt[digits[i]] = cnt.get(digits[i],0)+1
    if cnt[digits[i]] > 1:
        duplicate = True
    min_idx = dynamic_min_id(digits, i)
    if i == min_idx:
        continue
    else:
        digits[i], digits[min_idx] = digits[min_idx], digits[i]
        swapped = True
        break
if not swapped and not duplicate:
    digits[-1], digits[-2] = digits[-2], digits[-1]
print(*digits, sep="")
```

```python
# W9A6
def base(n):
    res = []
    for i in range(n):
        res.append(i+1)
    return res
def search(trace, pivot):
    for i in range(0,len(trace)):
        if trace[i] < pivot:
            return i-1
    return i
def nextperm(state):
    trace = [state[-1]]
    for i in range(len(state)-1,0,-1):
        if state[i-1]>state[i]:
            trace.append(state[i-1])
        else:
            j = search(trace, state[i-1])+i
            state[i-1], state[j] = state[j], state[i-1]
            state[i:] = sorted(state[i:])
            return state
    return base(len(state))
n = int(input())
a = list(map(int, input().split()))
print(nextperm(a))
```

```python
# W9A6.2
def nextbin(n, m, state, pivot):
    if len(pivot)<m:
        for i in range(len(state)-1, pivot[-1], -1):
            if state[i]==1 and state[i-1]==0:
                state[i], state[i-1] = state[i-1], state[i]
                pivot.append(i-1)
                print(*state, sep="")
                return None
    if len(pivot)>1:
        for i in range(m-1,0,-1):
            if state[pivot[i]-1]==1:
                pivot.pop()
            else:
                count = 0
                state[pivot[i]-1] = 1
                for j in range(len(state)-1, pivot[i]-1, -1):
                    if count < m-(i+1): 
                        state[j] = 1
                        count += 1;
                    else:
                        state[j] = 0
                pivot[i] -= 1
                print(*state, sep="")
                return None
    if len(pivot)==1:
        if pivot[0]==0:
            pivot.pop()
        else:
            count = 0
            state[pivot[0]-1] = 1
            for j in range(len(state)-1, pivot[0]-1, -1):
                if count < m-1: 
                    state[j] = 1
                    count += 1;
                else:
                    state[j] = 0 
            pivot[0] -= 1
            print(*state, sep="")
def bingen(n, m):
    state = [0]*n+[1]*m
    pivot = [n]
    print(*state, sep="")
    while len(pivot)>0:
        nextbin(n, m, state, pivot)
n, m = map(int, input().split())
bingen(n, m)
```

```python
# W9A7
def canPlaceFlowers(flowerbed, k):
    length = 0
    capacity = 0
    i = 0
    while i<len(flowerbed):
        if flowerbed[i]=="1":
            if length > 0:
                capacity += length//2
                length = 0
            if length == 0:
                length = -1
        else:
            length += 1
        i += 1
    capacity += (length+1)//2
    return (capacity >= k)
flowerbed = input().split()
k = int(input())
print(canPlaceFlowers(flowerbed, k))
```

```python
# W9A8
def is_path_crossing(moves):
    path = {(0,0): 1}
    pos = (0,0)
    for move in moves:
        if move=="R":
            pos = (pos[0]+1, pos[1])
        elif move=="L":
            pos = (pos[0]-1, pos[1])
        elif move=="U":
            pos = (pos[0], pos[1]+1)
        elif move=="D":
            pos = (pos[0], pos[1]-1)
        if path.get(pos, 0):
            return True
        else:
            path[pos] = 1
    return False
moves = input()
print(is_path_crossing(moves))
```

```python
# W9A10
def distance(house, heaters):
    left = 0
    right = len(heaters)-1
    while right-left>1:
        mid = (left+right)//2
        if house == heaters[mid]:
            return 0
        elif house < heaters[mid]:
            right = mid
        else:
            left = mid
    else:
        return min(abs(house-heaters[left]), abs(heaters[right]-house))
    
def findRadius(houses, heaters):
    heaters.sort()
    radius = 0
    for house in houses:
        radius = max(radius, distance(house, heaters))
    return radius
houses = list(map(int, input().split()))
heaters = list(map(int, input().split()))
print(findRadius(houses, heaters))
```
