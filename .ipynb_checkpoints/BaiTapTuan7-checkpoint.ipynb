```python
# W7A1
def binary_search(arr, left, right, target):
    if left > right:
        return -1
    mid = (left+right)//2
    if arr[mid] == target:
        return mid
    elif arr[mid] < target:
        left = mid+1
    else:
        right = mid-1
    return binary_search(arr, left, right, target)
a = list(map(int, input().split()))
x = int(input())
print(binary_search(a,0,len(a)-1,x))
```

```python
# W7A2
def count_occurences(arr, x):
    count = 0
    for e in arr:
        if e == x:
            count += 1
    return count

a = list(map(int, input().split()))
x = int(input())
print(count_occurences([],1))
```

```python
# W7A3
def bubble_sort(a):
    while True:
        swapped = False
        for i in range(len(a)-1):
            if a[i] > a[i+1]:
                a[i+1], a[i] = a[i], a[i+1]
                swapped = True
        if not swapped:
            break

def selection_sort(a):
    for i in range(len(a)):
        min_idx = i
        for j in range(i+1,len(a)):
            if a[j] < a[min_idx]:
                min_idx = j
        a[i], a[min_idx] = a[min_idx], a[i]  
a = list(map(int, input().split()))
b = a.copy()
print(b)
bubble_sort(a)
selection_sort(b)
print(a)
print(b)
```

```python
# W7A4
def first_mode(a):
    cnt = {}
    mode_key = a[0]
    for i in range(len(a)):       
        if cnt.get(a[i]) == None:
            cnt[a[i]] = [i,1]
        else:
            cnt[a[i]][1] += 1
        if cnt[a[i]][1] > cnt[mode_key][1]:
            mode_key = a[i]
        elif cnt[a[i]][1] == cnt[mode_key][1]:
            if cnt[a[i]][0] < cnt[mode_key][0]:
                mode_key = a[i]
    return mode_key

a = list(map(int, input().split()))
print(first_mode(a))
```

```python
# W7A10
a = eval(input())
m = int(input())
prefix_sum = [0]
for i in range(len(a)):
    prefix_sum.append((prefix_sum[i]+a[i])%m)
_max = 0
for i in range(len(a)):
    for j in range(i,len(a)):
        _sum = (prefix_sum[j+1]-prefix_sum[i])%m
        if _sum > _max:
            _max = _sum
print(_max)
```

```python
# W7A13
a = list(map(int, input().split()))
n = len(a)
max_length = 1
for i in range(n):
    found = [0, 0]
    length = [1,1]
    for j in range(i+1,n):
        if a[j]-a[i]==0:
            length[0] += 1
            length[1] += 1
        elif a[j]-a[i]==1:
            length[0] += 1
            found[0] = 1
        elif a[j]-a[i]==-1:
            length[1] += 1
            found[1] = 1
    length[0] = length[0]*found[0]
    length[1] = length[1]*found[1]
    max_length = max(max_length, max(length))
print(max_length)
```
