```python
# W6A1 
def mark(lst):
    dct = {}
    for e in lst:
        if dct.get(e,0)==0:
            dct[e] = 1
    return dct

lst = input().split()
for e in mark(lst):
    if mark(lst)[e] == 1:
        print(e)
```

```python
# W6A2
lst = list(map(int, input().split()))
cum = [0]
for i in range(len(lst)):
    cum.append(cum[i]+lst[i])
cum.pop(0)
for i in range(len(lst)):
    print(cum[i], end=" ")
```

```python
# W6A3
tup = tuple(input().split())
k = int(input())

l = len(tup)
new = list(tup)
for i in range(l):
    new[(i-k)%l]=tup[i]
new = tuple(new)

print(new)
```

```python
# W6A4
lst = input().split()
dct = {}
for e in lst:
    key, value = e.split(":")
    if dct.get(key,0)==0:
        dct[key] = []
    dct[key].append(value)
print(dct)
```

```python
# W6A5
lst = list(map(int, input().split()))
dct = {'positives':0, 'negatives':0, 'zeros':0}
for e in lst:
    if e > 0:
        dct['positives'] += 1
    elif e < 0:
        dct['negatives'] += 1
    else:
        dct['zeros'] += 1
print(dct)
```

```python
# W6A7
tup = tuple(map(int, input().split()))
rev_tup = list(tup)
l = 0
r = len(tup)-1
while l<r:
    rev_tup[l], rev_tup[r] = rev_tup[r], rev_tup[l]
    l += 1
    r -= 1
rev_tup = tuple(rev_tup)
print(tup[0], tup[-1], rev_tup)
```

```python
# W6A9 (pending)
def readDict(s):
    lst = s.split()
    dct = {}
    for e in lst:
        key, value = e.split(":")
        dct[key] = value
    return dct

a = readDict(input())
b = readDict(input())
for key in b.keys():
    if a.get(key, "")=="":
        a[key] = b[key]
    else:
        a[key] += b[key]
for key in sorted(a.keys()):
    print(key, a[key], end=" ")
```

```python
# W6A10 
lst = list(map(int, input().split()))
k = int(input())
pairs_dict = {}
for i in range(len(lst)):
    for j in range(i,len(lst)):
        if lst[i]+lst[j]==k:
            pairs_dict[min(lst[i],lst[j])]=max(lst[i],lst[j])
pairs_list = []
for key in sorted(pairs_dict.keys()):
    pairs_list.append((key, pairs_dict[key]))
print(pairs_list)
```

```python
# W6A11
tup = tuple(map(int, input().split()))
even = []
odd = []
for e in tup:
    if e%2==0:
        even.append(e)
    else:
        odd.append(e)
even = tuple(even)
odd = tuple(odd)
print(even)
print(odd)
```

```python
# W6A12
def cnt(lst):
    dct = {}
    max_count = 0
    for n in lst:
        dct[n] = dct.get(n, 0) + 1
        if dct[n] > max_count:
            max_count = dct[n]
            max_key = n
        elif dct[n] == max_count:
            if n < max_key:
                max_key = n
    return max_key
lst = list(map(int, input().split()))
print(cnt(lst))
```

```python
# W6A13
def readDict(s):
    lst = s.split()
    dct = {}
    for e in lst:
        key, value = e.split(":")
        dct[key] = value
    return dct

dct = readDict(input())
new = {}
for (key, value) in dct.items():
    new[value] = key
print(new)
```

```python
# W6A14
def mark(lst):
    dct = {}
    for e in lst:
        if dct.get(e,0)==0:
            dct[e] = 0
        dct[e] = 1
    return dct
    
lst1 = list(map(int, input().split()))
lst2 = list(map(int, input().split()))
dct1 = mark(lst1)
dct2 = mark(lst2)
lst = []
for key in dct1.keys():
    if dct2.get(key, 0) == 1:
        lst.append(key)
print(lst)
```

```python
# W6A15
def readDict(s):
    lst = s.split()
    dct = {}
    for e in lst:
        key, value = e.split(":")
        dct[int(key)] = value
    return dct
    
dct = readDict(input())
k = int(input())
dct_k = {}
for (key, value) in dct.items():
    if key > k:
        dct_k[key] = value
print(dct_k)
```
