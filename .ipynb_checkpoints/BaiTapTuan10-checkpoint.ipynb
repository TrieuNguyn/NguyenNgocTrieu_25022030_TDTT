```python
# W9A1
a = list(map(int, input().split()))
b = list(map(int, input().split()))
i = j = 0
c = []
while i<len(a) and j<len(b):
    if a[i]<b[j]:
        c.append(a[i])
        i += 1
    else:
        c.append(b[j])
        j += 1
c.extend(a[i:])
c.extend(b[j:])
print(c)
```

```python
# W9A2
n, m = map(int, input().split())
a = []
for _ in range(n):
    a.append(list(map(int, input().split())))
k = int(input())
a_diags_flattened = []
l = 0
while l<m+n-1:
    for i in range(max(l-(m-1),0),min(n,l+1)):
        a_diags_flattened.append(a[i][l-i])
    l += 1
print(a_diags_flattened[k-1])
```

```python
# W9A6
def base(n):
    res = []
    for i in range(n):
        res.append(i+1)
    return res
def search(trace, pivot):
    for i in range(0,len(trace)):
        if trace[i] < pivot:
            return i-1
    return i
def nextperm(state):
    trace = [state[-1]]
    for i in range(len(state)-1,0,-1):
        if state[i-1]>state[i]:
            trace.append(state[i-1])
        else:
            j = search(trace, state[i-1])+i
            state[i-1], state[j] = state[j], state[i-1]
            state[i:] = sorted(state[i:])
            return state
    return base(len(state))
n = int(input())
a = list(map(int, input().split()))
print(nextperm(a))
```
