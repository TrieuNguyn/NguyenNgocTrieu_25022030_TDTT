```python
# W6A1 
def mark(lst):
    dct = {}
    for e in lst:
        if dct.get(e,0)==0:
            dct[e] = 1
    return dct

lst = input().split()
for e in mark(lst):
    if mark(lst)[e] == 1:
        print(e)
```

```python
# W6A2
lst = list(map(int, input().split()))
cum = [0]
for i in range(len(lst)):
    cum.append(cum[i]+lst[i])
cum.pop(0)
for i in range(len(lst)):
    print(cum[i], end=" ")
```

```python
# W6A3
tup = tuple(input().split())
k = int(input())

l = len(tup)
new = list(tup)
for i in range(l):
    new[(i-k)%l]=tup[i]
new = tuple(new)

print(new)
```

```python
# W6A4
lst = input().split()
dct = {}
for e in lst:
    key, value = e.split(":")
    if dct.get(key,0)==0:
        dct[key] = []
    dct[key].append(value)
print(dct)
```

```python
# W6A5
lst = list(map(int, input().split()))
dct = {'positives':0, 'negatives':0, 'zeros':0}
for e in lst:
    if e > 0:
        dct['positives'] += 1
    elif e < 0:
        dct['negatives'] += 1
    else:
        dct['zeros'] += 1
print(dct)
```

```python
# W6A6
lst = list(map(int, input().split()))
s = 0
for e in lst:
    s += e
print(s)
```

```python
# W6A7
tup = tuple(map(int, input().split()))
rev_tup = list(tup)
l = 0
r = len(tup)-1
while l<r:
    rev_tup[l], rev_tup[r] = rev_tup[r], rev_tup[l]
    l += 1
    r -= 1
rev_tup = tuple(rev_tup)
print(rev_tup)
```

```python
# W6A8
lst = input().split()
dct = {}
for s in lst:
    if dct.get(s,0) == 0:
        dct[s] = 0
    dct[s] += 1
print(dct)
```
